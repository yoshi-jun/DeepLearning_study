### 6.1.2 テンソルの計算

実際にPyTorchを使用する。tensorは多次元配列を扱うクラスで、Pytorchのように定義できる。

````python
x = torch.tensor(5.0, requires_grad = True)

y = 3 * x ** 2

print("x = ",x,"y = ", y)
````

これは、$1\times1$の行列を作成し計算を行った。次に、これを微分する。$y$は

$$y=3x^2$$

なので、その微分は

$$\frac{dy}{dx}=6x$$

である。この時、$x=5$では$y=30$となる。実際にbackward()メソッドを使用し計算すると

````python
y.backward()
print(x.grad)
````

となる。backward()メソッドでは、微分を計算し、その結果はx.gradに記録される。

### 6.1.3 勾配法

PyTorchを使用して簡単問題を解く。$x$の関数$y$が次の式で表される時の最小値を求める。

$$y = 100(x_1-x_0^2)^2+(x_0-1)^2$$

この関数はローゼンブロック関数と呼ばれ、最小値の探索が難しいため、最適化問題のベンチマークとして使用される。

はじめに、解析的に解くために$x_0,x_1$の微分を計算すると次のようになる。

$$\frac{dy}{dx_0} = 400x_0^3-400x_0x_1+2x_0-2=0$$
$$\frac{dy}{dx_1} = 200x_1-200x_0^2 = 0$$

この連立方程式を解くと$(x_0,x_1)=(1,1)$となり、最小値が$(1,1)$にあることがわかる。

````python
def rosenbrock(x0, x1):
    y = 100 * (x1 - x0 ** 2) ** 2 + (x0 - 1) ** 2
    return y

x0 = torch.tensor(0.0, requires_grad=True)
x1 = torch.tensor(2.0, requires_grad=True)

y = rosenbrock(x0, x1)
y.backward()

print(x0.grad, x1.grad)
````

````python
import numpy as np
import matplotlib.pyplot as plt

x0s = np.arange(-2, 2, 0.1)
x1s = np.arange(-2, 2, 0.1)

X, Y = np.meshgrid(x0s, x1s)
Z = 100 * (Y - X ** 2) ** 2 + (X - 1) ** 2

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x0')
ax.set_ylabel('x1')
ax.set_zlabel('y')
plt.show()
````

$x_0 = 0, x_1 = 2$の場合、解析的に解いた式に代入すると$(-2,400)$が得られる。これを、勾配という。

勾配は、関数の出力を**増加**させる方向を示す。つまり、勾配にマイナスをかけた値$(2,400)$が最も減らす方向であることを意味する。

勾配法をローゼンブロック関数に適用する。

````python
x0 = torch.tensor(0.0, requires_grad=True)
x1 = torch.tensor(2.0, requires_grad=True)

x0_grad = []
x1_grad = []

lr = 0.001
iters = 10000

for i in range(iters):
    if i % 1000 == 0:
        print(x0.item(), x1.item())
        
    y = rosenbrock(x0, x1)
    
    y.backward()
    
    #値の更新
    
    x0.data -= lr * x0.grad.data
    x1.data -= lr * x1.grad.data
    
    x0.grad.zero_()
    x1.grad.zero_()
    
    x0_grad.append(x0.item())
    x1_grad.append(x1.item())
print(x0.item(), x1.item())
````

````python
x0s = np.arange(-2, 2, 0.1)
x1s = np.arange(-1, 3, 0.1)

X, Y = np.meshgrid(x0s, x1s)

Z = 100 * (Y - X ** 2)** 2 + (X - 1) ** 2

ax = plt.axes()
ax.contour(X, Y, Z)
ax.grid()
plt.show()
````

更新する回数はitersとして設定する。lrはlearning lateの略称

テンソルの値の更新は、x0.data -= ler * x0.grad.data で実行する。

実際のデータに関して処理する場合には、x0.dataまたは、x0.garad.dataを操作する。基本的にPyTorchでは、値を更新する場合data属性を操作する。

item()メソッドは、スカラを持つtensorに対して使用することができる。通常は、損失関数や評価指標の値のログに記録したり、出力する場合に使用する。

スタート地点から徐々に更新され正解に近い解が得られることがわかる。

## 6.2 線形回帰

データから解法を見つけることが機械学習の本質である。

PyTorchを使用して機械学習の問題を解決する。

### 6.2.1 トイ・データセット

PyTorchを使用して問題を解くための小さなデータセットを作成する。この小さなデータセットはトイ・データセット(Toy Dataset)と呼ばれる。ここでは次の式で表されるデータセットを作成する。

$$y = 2x+ 5 + \epsilon$$

この式で$\epsilon$はノイズを意味し、$[0:1]$の一様分布から生成する。ノイズを追加する理由は、実問題においてもノイズが問題となり心地が計算できないことが多くある。

````python
import torch
torch.manual_seed(0)

x = torch.rand(100,1)
y = 2 * x + 5 + torch.rand(100,1)
````

````python
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

グラフを見ると、$x,y$はそれぞれ線形関係であることが観測できる。

### 6.2.2 線形回帰の理論

$y$と$x$が線形であると仮定している。$x$から予測される$y$を$\hat{y}$と表記すると次の式で表される。

$$\hat{y} = Wx+ b$$

ここで、$W$と$b$はスカラのパラメータとする。一方で、データにノイズが含まれているため誤差$y-\hat{y}$は正規分布に従う。

具体的には、平均$0$であり標準偏差$\sigma$の正規分布に従う。これは、$y$が平均$\hat{y}$、標準偏差$\sigma$に従うということである。これを数式で表すと

$$\hat{y} = Wx + b $$
$$p(y|x;w,b) = \mathcal{N}(y;\hat{y},\sigma)$$

ここで、$p(y|x;w,b)$の対数尤度は

$$\log p(y|x;w,b) = \log\mathcal{N}(y;\hat{y},\sigma)\\
= \log\left(\frac{1}{\sqrt{2\pi\sigma^2}}\exp\left(-\frac{(y-\hat{y})^2}{2\sigma^2}\right)\right)\\
= -\frac{1}{2\sigma^2}(y-\hat{y})^2+\log\frac{1}{\sqrt{2\pi\sigma^2}}\\
= -\frac{1}{2\sigma^2}(y-Wx-b)^2+\log\frac{1}{\sqrt{2\pi\sigma^2}}
$$

と表すことができる。パラメータ$W, b$が含まれる項を目的関数とすると、目的関数は

$$ -\frac{1}{2\sigma^2}(y-Wx-b)^2$$

となる。さらに、目的関数にマイナスをつけることで、最小化する問題として扱うことができる。

$$\arg\max_{W,b}\log p(y|x;W,b)=\arg\max_{W,b}\left(-\frac{1}{2\sigma^2}(y=Wx+b)^2\right)\\
= \arg\max_{W,b}(-(y-Wx-b)^2)\\
= \arg\min_{W,b}(y-(Wx+b))^2$$

これにより、$y$と$\hat{y} = Wx+b$の二乗誤差を最小化する問題である。実際には、データが$N$個あるため次の式で表される。

$$L(W,b)=\frac{1}{N}\sum^N_{n=1}\left(y^{(n)}-\left(Wx^{(n)}+b\right)\right)$$

関数$L(W,b)$は損失関数(Loss function)とも呼ばれる。ここで、は二乗誤差の平均を求めている。これを平均二乗誤差(Mean Squared Error)と呼ぶ。

この問題は、解析的に解くこともできるが、勾配法を用いて解いていく。

### 6.2.3 線形回帰の実装

PyTorchを用いて線形回帰を実装する。

````python
import torch

torch.manual_seed(0)

x = torch.rand(100, 1)
y = 5 + 2 * x + torch.rand(100, 1)

W = torch.zeros((1, 1), requires_grad = True)
b = torch.zeros(1, requires_grad = True)

def predict(x):
    y = x @ W + b   # 行列積の演算＠を使用　
                    # (100, 1) x (1, 1) = (100, 1)
                    # (100, 4) x (4, 1) = (100, 1)
                    # x.shape[1], W.shape[0]を一致させることで正しく計算が行われる。
    return y

def mean_squared_error(x0, x1): # 二乗平均を求める関数　torchにはmse_loss関数が用意されている。
    diff = x0 - x1
    N = len(diff)               # 要素数Nをカウント　
    return torch.sum(diff ** 2) / N

lr = 0.1 # 学習レート
iters = 100

for i in range(iters):
    y_hat = predict(x) # yハットを計算
    loss = mean_squared_error(y, y_hat) # 二乗平均誤差を計算

    loss.backward()

    W.data -= lr * W.grad.data # 傾きx学習レート
    b.data -= lr * b.grad.data

    W.grad.zero_()
    b.grad.zero_()

    if i % 10 == 0: # print every 10 iterations
        print(loss.item())

print(loss.item())
print('====')
print('W =', W.item())
print('b =', b.item())

# 描画

import matplotlib.pyplot as plt
x_learnd = np.arange(0,1,0.1)
y_learnd = [x * W.item() + b.item() for x in x_learnd]
plt.scatter(x, y)
plt.plot(x_learnd, y_learnd, color = 'red')
````

## 6.3 パラメータとオプティマイザ

ニューラルネットワークの学習には多くのパラメータを用いるため、自動管理する仕組みが用意されている。

* torch.nn.Parameterクラス
* torch.nn.Moduleクラス
* torch.optim.SGDクラス

上記のクラスを使用し、前章で作成したプログラムを記述し直す。

### 6.3.1 ParameterとModuleクラス

toch.nn.ParameterクラスはTensorを継承したクラス。次のように使用する。

````python
import torch
import torch.nn as nn

W = nn.Parameter(torch.zeros(1, 1))
b = nn.Parameter(torch.zeros(1))

print(W)
print(b)
````

Parameterはデフォルトでrequire_grad=Trueで作成されていることが確認できる。

ParameterクラスはModuleクラスと一緒に使うことで威力を発揮する。

````python
class Model(nn.Module): # nn.Module クラスを継承する
    def __init__(self):
        super().__init__()
        self.W = nn.Parameter(torch.zeros(1, 1)) # nn.Parameterクラスのインスタンス
        self.b = nn.Parameter(torch.zeros(1))
        
    def forward(self, x): # 線形回帰の計算
        y = x @ self.W + self.b
        return y
    
model = Model()

for param in model.parameters():
    print(param)
````

PyTorchで線形変換を行うtorch.nn.Linearクラスがある。線形変換とは、$y = Wx +b$の計算を意味する。

````python
class Model(nn.Module): # nn.Module クラスを継承する
    def __init__(self, input_size=1, output_size = 1):
        super().__init__()
        self.linear = nn.Linear(input_size, output_size) # Linearのクラスを使用
        
    def forward(self, x): # 線形回帰の計算
        y = self.linear(x) #以前は、y = x @ self.W + self.b
        return y
    
model = Model()

for param in model.parameters():
    print(param)
````

Linearクラスの引数input_size, output_sizeはそれぞれ、入力するベクトルの次元数と出力するベクトルの次元数を意味する。

インスタントのパラメータはself.linear.b, self.linear.Wとしてアクセスすることができる。
また、自動的にModuleのパラメータに登録されるためmodel.parameters()で取得することができる。

### 6.3.2 オプティマイザ

オプティマイザは最適化手法を具体的に実装したものである。

前節で実装したコードは、次のように実装できる。

````python
x = torch.rand(100, 1)
y = 5 + 2 * x + torch.rand(100, 1)

lr = 0.1
iters = 100

model = Model()
optimizer = torch.optim.SGD(model.parameters(), lr = lr) # オプティマイザの生成と学習率を設定する

for i in range(iters):
    y_hat = model(x)
    loss = nn.functional.mse_loss(y, y_hat)
    
    loss.backward()
    optimizer.step() # パラメータの更新を行う
    optimizer.zero_grad() # 勾配をリセットする
````

PyTorchのtorch.optimには他にも様々な最適化方法がある。今回は、SGDを使用したがAdamなどもある。

### 6.4.1 非線形なデータセット

前回よりも複雑なデータセットとして次の式をデータセットとする。

$$ y = \sin(2\pi x) + \epsilon$$

$\epsilon$は一様乱数とする。このデータセットは次のコードで生成できる。

````python
import torch

torch.manual_seed(0)

x = torch.rand(100, 1)
y = torch.sin(2 * torch.pi * x) + torch.rand(100, 1)

#　描画

import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

$x,y$は線形の関係ではなく非線形である。線形回帰では対応できないため、ニューラルネットワークを使用する。

### 6.4.2 線形変換と活性関数

ニューラルネットワークでは線形の出力に対して非線形な変換を行う。その関数を活性化関数と呼ぶ。代表的なものとしてReLu,シグモイド関数がある。

シグモイド関数は
$$y = \frac{1}{1+e^{-x}}$$
と表される。
ReLU関数は
$$y = \left\{
\begin{array}{ll}
0 & (x<0) \\
x & (x\geq0)
\end{array}
\right.$$
と表される。描画すると以下のようになる。

````python
x = np.arange(-5,5,0.1)

def sigmoid(x):
    y = 1 / (1+np.exp(-x)) 
    return y

def relu(x):
    N = len(x)
    y = np.zeros(N)
    
    for i in range(N):
        if x[i] < 0:
            y[i] = 0
        else:
            y[i] = x[i]
    return y

y_sig = sigmoid(x)
y_rel = relu(x)

fig = plt.figure(figsize=[10,4])
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_title('Sigmoid')
ax1.grid()
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.plot(x, y_sig)

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_title('ReLu')
ax2.grid()
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.plot(x, y_rel)
plt.show()
````

両方とも非線形な関数であり、これがテンソルの要素ごとに適用される。

torch.nn.functionalを用いて次のように使用することができる。

````python
import torch.nn.functional as F

a = torch.arange(-5,5, 0.1)

b = F.sigmoid(a)
c = F.relu(a)

plt.scatter(a,b, label = 'sigmoid')
plt.scatter(a,c, label = 'relu')
plt.legend()
plt.grid()
plt.show()
````

### 6.4.3 ニューラルネットワークの実装

一般的なニューラルネットワークは線形変換と活性化関数を交互に使用する。

例として二層のニューラルネットワークを実装すると次のようになる。

````python
import torch
import torch.nn as nn
import torch.nn.functional as F


torch.manual_seed(0)
x = torch.rand(100, 1)
y = torch.sin(2 * torch.pi * x) + torch.rand(100, 1)

# model
class Model(nn.Module): #Moduleクラスを継承して作成
    def __init__(self, input_size=1, hidden_size= 10, output_size=1):
        super().__init__()
        self.linear1 = nn.Linear(input_size, hidden_size) # 線形へ印鑑を行う部分が２つに追加されている。
        self.linear2 = nn.Linear(hidden_size, output_size)# これは、ax+b, cx+dの二つがあるということ。パラメータは別

    def forward(self, x): #実際に計算を行っている部分。ここでは線形変換その１→活性化関数→線形変換その２で出力する。
        y = self.linear1(x)
        y = F.sigmoid(y)
        y = self.linear2(y)
        return y
````

実際に、ニューラルネットワークの学習を行う。

````python
lr = 0.2
iters = 10000
loss_hist = []

model = Model()# モデルのインスタンスを作成
optimizer = torch.optim.SGD(model.parameters(), lr=lr)#最適化関数の実装と学習率の設定

for i in range(iters):
    y_pred = model(x)                        # 現在の出力をy_predに記録
    loss = F.mse_loss(y, y_pred)             # yとy_predの二乗平均誤差を計算
    optimizer.zero_grad()                    # 勾配の初期化
    loss.backward()                          # バックプロぱゲーション（微分）を実行
    optimizer.step()
    
    loss_hist.append(loss.item())
    
    if i % 1000 == 0:
        print(loss.item())

print(loss.item())
````

損失関数計算結果lossを描画し、確認すると

````python
plt.plot(loss_hist)
plt.show()
````

lossの値は単調に減少しているわけではないが、減少していっていることが確認できる。

学習後のネットワークの出力を確認すると

````python
# plot
import matplotlib.pyplot as plt
plt.scatter(x.detach().numpy(), y.detach().numpy(), s=10)
x = torch.linspace(0, 1, 100).reshape(-1, 1)
y = model(x).detach().numpy()
plt.plot(x, y, color='red')
plt.show()
````

線形変換と活性化関数を重ねることで、非線形な関係でも学習できることが確認できた。

### 6.5.4 データローダ

学習に使用する膨大なデータを一度に使用することは困難である。そのため、データセットを分割する。分割したデータをミニバッチと呼ぶ。

torch.utils.data.DataLoaderを使用するとデータセットの操作が容易になる。

````python
# ==== DataLoader ====
dataloader = torch.utils.data.DataLoader( # データローダのインスタンスを作成
    dataset,                              # データセット
    batch_size=32,                        # ミニバッチの個数。つまり分割数
    shuffle=True)                         # エポックごとにデータを並べ替える

for x, label in dataloader:
    print('x shape:', x.shape)            # shape: torch.Size([32, 1, 28, 28])
    print('label shape:', label.shape)    # shape: torch.Size([32])
    break                                 # 0番目のミニバッチのみ表示するためbreak
````

出力を確認すると、[32, 1, 28, 28]の形状であり、ミニバッチのデータがテンソルにまとめられていることがわかる。同様にlabelの形状も[32]になっている。
