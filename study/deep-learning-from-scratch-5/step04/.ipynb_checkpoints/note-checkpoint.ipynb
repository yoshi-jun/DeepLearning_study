# 4 混合ガウスモデル

一つの正規分布では表せない現象を、複数の正規分布を組み合わせた複数の山で表現する。これにより、より複雑な表現が可能になる。これは、混合ガウスモデルと呼ばれる。

## 4.1 身の回りにある多峰性分布

複数の山を持つ分布、多峰性分布が身近に多く見られる。その例を示す。

**男女混合の身長分布** 

年齢と性別を限定した場合に、正規分布になる。性別を限定しない場合、性別ごとにピークが異なる。

**アリの体長**

アリは役割に応じて体の大きさが異なるため、アリの集団の全体を通してみると、体長の分布は複数の山から構成される。

**間欠泉の噴出**

一定周期で熱湯を噴出する温泉を間欠泉と呼ぶ。間欠泉の噴出には規則性があり、ここにも多様性の分布が見られる。

## 4.2 混合ガウスモデルのデータ生成

先に説明したように、多峰性分布は複数の正規分布を組み合わせることで表現することができる。

複数の正規分布からなるモデルは、**混合ガウスモデル(Gaussian Mixsture Model)** と呼ばれる。混合ガウスモデルは、GMMと表記される。

GMMによる生成モデルを作成する。モデル作成は先の例(身長の一次元正規分布、身長と体重の多次元正規分布)と同様に

1. モデル化 : 観測データの分布がGMMによって表現できると仮定する
2. パラメータ推定 : GMMのパラメータを推定する

この二つの手順で、GMMによる生成モデルを作成することができる。

しかし、先の例と異なり最尤推定によってパラメータを解析的に解くことができない。EMアルゴリズムといった手法を使用する必要がある。（EMアルゴリズムについては、Step5で説明する。）

ここでは、推定が終わり、パラメータが判明しているモデルから、データの生成を行う。

### 4.2.1 GMMによるデータ生成

ここでは、データ生成する手順について説明する。

簡単な例として、二つの正規分布による混合ガウスモデル(GMM)を考える。生成手順は、
  1. 二つの正規分布の中から、ある確率分布に従って一つを選ぶ
  2. 選んだ正規分布からデータを生成する。

これを繰り返し多峰分布を生成する。

### 4.2.2 データ生成を行うコード

実際にデータを生成する。生成するコードを下に示す。今回は、事前に推定したパラメータを使用する。

```python
import numpy as np

# 事前に推定したパラメータ。平均、分散行列が二つあることに注意

mus  = np.array([[2.0, 54.50],
                 [4.3, 80.0]])
covs = np.array([[[0.07, 0.44],
                  [0.44, 33.7]],
                 [[0.17, 0.94],
                  [0.94, 36.00]]])
phis = np.array([0.35, 0.65])

def sample():
    z = np.random.choice(2, p=phis)
    mu, cov = mus[z], covs[z]
    x = np.random.multivariate_normal(mu, cov)
    return x
```

mus, covsはそれぞれ推論済みのパラメータである。それぞれのshapeを確認すると

```python
print("mus.shape ", mus.shape)
print("covs.shape", covs.shape)
```

二つの正規分布の平均と、共分散行列がまとめられていることがわかる。

次に、phisは各正規分布が選ばれる確率を表す。phisの要素数は、選択する正規分布の数と等しい。

```python
print("phis  ", phis)
```

サンプリングはsamp()関数で行う。サンプリングの手順は、使用する正規分布をphisの確率で選択、選択した正規分布のパラメータ(平均ベクトル、共分散行列)を使用してサンプリングする。

```python
for _ in range(10):
    z = np.random.choice(2, p=phis)
    print(z)
```

phisに従って１または０を返す。

次に、選択した正規分布のパラメータmu, covを取得する。

```python
mu, cov = mus[z], covs[z]
```

```python
print("mu  ",mu)
print("cov ",cov)
```

正規分布のパラメータを決定し、多次元正規分布からデータを一つサンプリングする。

```python
np.random.multivariate_normal(mu, cov)
```

これを複数回繰り返すことで、データをサンプリングし、散布図を描画する。

```python
N = 500
xs = np.zeros((N,2))
for i in range(N):
    xs[i] = sample()
    
plt.scatter(xs[:,0], xs[:,1], color="orange", alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## 4.3 混合ガウスモデルの式

前節では、GMMによるデータ生成について確認した。GMMを数式で表すために、同時確率、条件付き確率について復習する。

### 4.3.1 確率の復習

確率変数$x,y$が同時に起こる確率$p(x,y)$は　**同時確率（Joint Probablity）**と呼ばれる。一方で、　**周辺確率(Marginal Probability)**　は、$p(x),p(y)$のように単独の確率変数が現れる確率を示す。

同時確率から、周辺確率を求めることができ、これを　**周辺化(Marginalization)**　と呼ぶ。yが離散変数の場合

$$ p(x) = \sum_y p(x,y) $$

連続関数の場合

$$ p(x) = \int p(x,y)dy $$

で周辺化することができる。

**条件付き確率(Conditional Probability)**　はある条件下での確率を表す。$y$が起きている状況のもとでの$x$の確率は、$p(x|y)$と表記する。

紹介した、同時確率、周辺確率、条件付き確率の三つの確率は下の、 **乗法定理**で関連づけることができる。

$$ p(x,y) = p(x|y)p(y) $$

### 4.3.2 GMMの式

GMMは次の手順でデータを生成することを前節で確認した。

K個の正規分布を用意する。
  1. K個の正規分布の中から、ある確率分布に従って一つを選ぶ
  2. 選んだ正規分布からデータを生成する

確率分布を選ぶステップは、**カテゴリカル分布(Categorical Distribution)** によって表される。カテゴリ軽分布は次の式で表すことができる。

$$ p(z=k; \phi) = \phi_k $$

離散型の確率変数$z$は1から$K$の間の整数$k$をとる。$\phi$は$K$個の値をもつ。

クラスが二つの場合は、$\phi = (0.35, 0.65)$の2つのパラメータを持つベクトルとなる。

カテゴリカル分布のパラメータ$\phi = (\phi_1, \phi_2, \cdots, \phi_K)$は次の条件を満たす。

  * すべての$k$において、$\phi_k \geq 0$
  * $\sum^K_{k=1} \phi_k = 1$
  
続いて、複数の正規分布を数式で表す。$K$個の正規分布のパラメータを次のように準備する。

$$ \boldsymbol{\mu} = \{\boldsymbol{\mu_1}, \boldsymbol{\mu_2}, \cdots, \boldsymbol{\mu}_K\} $$
$$ \boldsymbol{\Sigma} = \{\boldsymbol{\Sigma_1}, \boldsymbol{\Sigma_2}, \cdots, \boldsymbol{\Sigma}_K\} $$

全部で二つの正規分布がある場合、用意したパラメータを用いて、二つの正規分布
$\mathcal{N}(\boldsymbol{x},\boldsymbol{\mu}_1,\boldsymbol{\Sigma_1}), \mathcal{N}(\boldsymbol{x},\boldsymbol{\mu}_2,\boldsymbol{\Sigma_2})$が得られる。

ここでは、正規分布が$K$個用意された場合を考える。$z$が取る値によって、使われる正規分布が決まるとすると、次のように表すことができる。

$$p(\boldsymbol{x}| z = k; \boldsymbol{\mu},\boldsymbol{\Sigma}) = \mathcal{N}(\boldsymbol{x},\boldsymbol{\mu}_k,\boldsymbol{\Sigma}_k)$$

生成モデルの目標は、観測データ$x$の確率分布$p(x)$を表すことである。GMMの場合は、確率の周辺化により、$p(\boldsymbol{x},z)$を使用して$p(\boldsymbol{x})$を表すことができる。

$$p(\boldsymbol{x})=\sum^K_{k=1}p(\boldsymbol{x},z=k)$$

そして、同時確率$p(\boldsymbol{x},z=k)$は次の式で表すことができる。

$$p(\boldsymbol{x},z=k)=p(z=k)p(\boldsymbol{x}|z=k) \\
  =\phi_k\mathcal{N}(\boldsymbol{x},\boldsymbol{mu}_k,\boldsymbol{\Sigma}_k)$$

上の結果を代入すると、

$$p(\boldsymbol{x}) = \sum^K_{k=1}\phi_k\mathcal{N}(\boldsymbol{x};\boldsymbol{\mu}_k,\boldsymbol{\Sigma})$$

と記述することができる。この式がGMMの式である。各正規分布に対して重み$\phi$がかけられそれらを全て足し合わす。つまり、重み$\phi$を持つ正規分布の和である。

GMMでは、確率変数が2つあり、$\boldsymbol{x}$と$z$が存在する。$\boldsymbol{x}$は、学習用の観測データに対応する。一方で、$z$は観測されず、**潜在変数(Latent Variable)** という

### 4.3.3 GMMの実装

GMMを実際に実装する。

```python
import numpy as np
import matplotlib.pyplot as plt

mus = np.array([[2.0, 54.50],
                [4.3, 80.0]])

covs = np.array([[[0.07, 0.44],
                  [0.44, 33.7]],
                 [[0.17, 0.94],
                  [0.94, 36.00]]])

phis = np.array([0.35, 0.65])

def multivariate_normal(x, mu, cov):
    det = np.linalg.det(cov)
    inv = np.linalg.inv(cov)
    d   = len(x)
    
    z = 1 / np.sqrt((2 * np.pi) ** d * det)
    y = z * np.exp((x - mu).T @ inv @ (x - mu) / -2.0)
    
    return y

def gmm(x, phis, mus, covs):
    K = len(phis)
    y = 0
    for k in range(K):
        phi, mu, cov = phis[k], mus[k], covs[k]
        y += phi * multivariate_normal(x, mu, cov)
    return y
```

multivariate_normal()関数では、多次元の正規分布を出力する。

detは行列式、invは逆行列をNumPyのlinalg関数を使用しそれぞれ計算している。dは次元を表し、確率変数の次元を取得している。

zは、$\frac{1}{\sqrt{(2\pi)^D|\Sigma|}}$を計算している。これは、$\mathcal{N}$の係数である。

yは、$z\exp{\left\{-\frac{1}{2}(\boldsymbol{x}-\boldsymbol{\mu})^\top\Sigma^{-1}(\boldsymbol{x}-\boldsymbol{\mu})\right\}}$を計算し、$\mathcal{N}$を計算している。

次にGMMの計算である。gmm関数で計算を行っている。

Kは入力している正規分布を選択する確率phisの長さで、正規分布の数を意味する。

次に、forループを回し、各正規分布ごとにphi, mu, covの値を取得する。
それらのパラメータを用いて、多次元正規分布からサンプリングし、phiと掛け合わせyに加算する。

実際に可視化すると次のようになる。

```python
xs = np.arange(1, 6, 0.1)
ys = np.arange(40, 100, 0.1)
X, Y = np.meshgrid(xs, ys)
Z = np.zeros_like(X)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        x = np.array([X[i, j], Y[i, j]])
        Z[i, j] = gmm(x, phis, mus, covs)

fig = plt.figure(figsize=[20,10])
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
ax1.plot_surface(X, Y, Z, cmap='viridis')

ax2 = fig.add_subplot(1, 2, 2)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.contour(X, Y, Z)
ax2.grid()
plt.show()
```

それぞれのピークの高さが異なっていることがわかる。これは、phiに依存しており、phiが大きいほど、高くなる。

## 4.4 パラメータ推定の難所

GMMについて学んできたが、パラメータ推定について述べる。

GMMのパラメータ推定が難点と言われる所以を数式を用いて示す。

### 4.4.1 GMMのパラメータ推定

GMMは次の式で示される。

$$p(\boldsymbol{x};\phi,\boldsymbol{\mu},\boldsymbol{\Sigma})=\sum^K_{k=1}\phi_k\mathcal{N}(\boldsymbol{x};\boldsymbol{\mu}_k,\boldsymbol{\Sigma}_k)$$

ここで、あるサンプル$\mathcal{D}=(\boldsymbol{x}^{(1)},\boldsymbol{x}^{(2)},\cdots,\boldsymbol{x}^{(k)}$が得られた場合を考える。パラメータ$\{\phi,\boldsymbol{\mu},\boldsymbol{\Sigma}\}$をまとめて$\boldsymbol{\theta}$と表すと

$$p(\mathcal{D};\boldsymbol{\theta} = p (\boldsymbol{x}^{(1)};\boldsymbol{\theta})p(\boldsymbol{x}^{(2)};\boldsymbol{\theta})\cdots p(\boldsymbol{x}^{(N)};\boldsymbol{\theta}) = \prod^N_{n=1}p(\boldsymbol{x}^{(n)};\boldsymbol{\theta})$$

最優推定を行う。尤度$p(\mathcal{D};\boldsymbol{\theta})$を最大化するためのパラメータ$\theta$を見つける。計算を簡略化するために、対数尤度$\log p(\mathcal{D};\theta)$を最大化することを考える。

$$ \log p(\mathcal{D};\boldsymbol{\theta}) = \log \prod^N_{n=1}p(\boldsymbol{x}^{(n)};\boldsymbol{\theta}) = \sum^N_{n=1}\log p(\boldsymbol{x}^{(n)};\boldsymbol{\theta}) = \sum^N_{n=1} \log \left( \sum^K_{k=1}\phi_k\mathcal{N}(\boldsymbol{x}^{(n)};\boldsymbol{\mu}_k,\boldsymbol{\Sigma}_k)\right)
$$

正規分布の合計であり、傾きが$0$になる位置で対数尤度$L(\mathcal{D};\boldsymbol{\theta})=\log p(\mathcal{D};\boldsymbol{\theta})$は最大値になる。つまり、次の方程式を解く必要がある。

$$\frac{\partial L}{\boldsymbol{\partial\phi}} = 0,　\frac{\partial L}{\boldsymbol{\partial\mu}} = 0,　\frac{\partial L}{\boldsymbol{\partial\Sigma}} = 0$$

この式を解析的に計算を行うことはできない。今までは、sum-logの形であり、解析的に解くことが容易であった。だが、今回は式がlog-sumという形をしているため解くことは難しい。

$K=1$の場合は、logとexpが互いに打ち消しあい計算することができる。一方で、複数のlogがある場合は打ち消すことができず、解析的な解が得られない。

そのため、EMアルゴリズムによってパラメータを推定する。EMアルゴリズムについては、次のステップで学習する。
