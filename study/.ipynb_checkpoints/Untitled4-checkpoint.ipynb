モデルの評価

```python
import tensorflow as tf
import numpy as np
import pandas as pd

# Allocation GPU hardware ID = 0
physical_devices = tf.config.list_physical_devices('GPU')

tf.config.set_visible_devices(physical_devices[1], 'GPU')
```

```python
model = tf.keras.models.load_model("model_01")
```

```python
dir_name = "data_02" 
theta = 0
file_names = [str(dir_name)+"/data"+str(theta)+","+str(phi)+".csv" for phi in range(-45,45)]

data = [np.array(pd.read_csv(file_name)['dose']).reshape(61,61,150) for file_name in file_names]

data = np.array(data).astype("float32")
data = np.expand_dims(data, axis=4)
```

```python
pred = model.predict(data)
```

```python
import matplotlib.pyplot as plt
for k in range(len(pred)):
    
    fig = plt.figure()
    theta =[]
    phi = []
    
    for i in range(19):
        for j in range(19):
            theta.append(i)
            phi.append(j)
    
    d_theta = [1 for i in range(19*19)]
    d_phi = [1 for i in range(19*19)]
    
    # plt.imshow(pred.reshape(15,19,19)[i])
    ax = fig.add_subplot(projection='3d')
    ax.bar3d(theta,phi,0,d_theta,d_phi,pred.reshape(90,361)[k])
    plt.show()
    plt.close()
    
    print(pred.reshape(90,361)[k].sum())
```
