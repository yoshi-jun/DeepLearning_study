# Functional API を使用して

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Allocation GPU hardware ID = 0
physical_devices = tf.config.list_physical_devices('GPU')
tf.config.set_visible_devices(physical_devices[1], 'GPU')
```

classで多出力モデルを作成している例が見つからなかったため、Functional APIを用いてクラスを作成した。

```python
#Input shape の定義
inp_shape = (61, 61, 150, 1)
inputs = tf.keras.Input(shape = inp_shape)

# x = layer()(x)の形で層を定義していく
x = tf.keras.layers.Conv3D(16, 2, activation = 'relu', strides = 2, name = "conv3d_1")(inputs)
x = tf.keras.layers.BatchNormalization(name = "normalize_1")(x)

x = tf.keras.layers.Conv3D(32, 2, activation = 'relu', strides = 2, name = "conv3d_2")(x)
x = tf.keras.layers.BatchNormalization(name = "normalize_2")(x)

x = tf.keras.layers.Conv3D(64, 2, activation = 'relu', strides = 2, name = "conv3d_3")(x)

x = tf.keras.layers.Flatten(name = "flatten")(x)

# 出力層を複数にする場合は、出力層 = layer()(中間層)の形にする。
theta = tf.keras.layers.Dense(91, activation='softmax', name = "theta")(x)
phi = tf.keras.layers.Dense(91, activation='softmax'  , name = "phi"  )(x)

#ここで、modelのinputとoutputを定義する
model = tf.keras.Model(inputs=inputs, outputs=[theta, phi])

#出力層ごとにoptimizerやlossを変更できるが、全て同じで良い場合はひとつ設定すれば良い
model.compile(loss      = tf.keras.losses.BinaryCrossentropy(),
              optimizer = tf.keras.optimizers.SGD(0.2))

model.summary()
```

推論する粒度を細かくしたが、学習するdenseのパラメータの数が半分になっている。学習の進みが早くなる事が期待できる。

れだけでも、この構造にした意味をこ感じられる。

```python
x_train = np.load("data.npy")
x_train = np.expand_dims(x_train, axis=4)
t_train = np.load("theta.npy")
# t_train = np.expand_dims(t_train, axis=2)
p_train = np.load("phi.npy")

print(type(x_train))
```

# 学習

```python
model.fit(x = x_train, y = {"theta":t_train, "phi":p_train},
          batch_size = 1400,
          epochs     = 900)
```

# modelの性能を測る

```python
dir_name = "data_05" 
theta = 0
file_names = [str(dir_name)+"/"+str(theta)+","+str(phi)+".csv" for phi in range(-45,46)]

data = [np.array(pd.read_csv(file_name)['dose']).reshape(61,61,150) for file_name in file_names]

data = np.array(data).astype("float32")
data = np.expand_dims(data, axis=4)
```

```python
model = tf.keras.models.load_model("model_two_output")
pred_t, pred_p = model.predict(data)
```

```python
print(pred_t.shape)
t = [-45 + i for i in range(91)]
for i in range(91):
    plt.scatter(t,pred_p[i], color = "blue"  , label = "phi")
    plt.axvline(-45+i)
    plt.scatter(t,pred_t[i], color = "orange", label = "theta" )
    plt.axvline(0,color = "orange")
    plt.legend()
    plt.grid()
    plt.show()
    plt.close()
```

推論の結果の値が大きくなっている。これは、one-hotの行列の数が減った事が要因か？

ただ、1度間隔では表示されていない。

## ピーク

```python
peak_t = [np.argmax(pred_t[i]) - 45 for i in range(91)]
peak_p = [np.argmax(pred_p[i]) - 45 for i in range(91)]
```

```python
diff_p = [((-45 + i) - peak_p[i]) for i in range(91)]
diff_t = [0 -(peak_t[i])          for i in range(91)]
```

```python
plt.scatter(t, diff_p)
plt.plot(t,diff_p)
# plt.scatter(t, diff_t)
plt.xlabel("theta [degree]")
plt.ylabel("diff [degree]")
plt.grid()
```

ピークで考えると結果はこのようになる。5の倍数の角度以外の確率がほとんどゼロなため、ピークで取るとこうなる。（これは前回比較していない）

4を超える位置での誤差が見られる。

前回同様に重み付き平均で考えてみる。

## 重み付き平均

```python
mean_p = []
mean_t = []
for i in range(91):
    m_t = 0
    m_p = 0
    for j in range(91):
        idx_t = j - 45
        m_t += idx_t * pred_t[i][j]
        m_p += idx_t * pred_p[i][j]
    
    mean_t.append(m_t)
    mean_p.append(m_p)
```

```python
diff_p = [((-45 + i) - mean_p[i]) for i in range(91)]
diff_t = [0 -(mean_t[i])          for i in range(91)]
```

```python
plt.scatter(t, diff_p, label = "phi"  , s = 5)
plt.scatter(t, diff_t, label = "theta", s = 5)
plt.xlabel("theta [degree]")
plt.ylabel("diff [degree]")
plt.ylim(-2.5,2.5)
plt.legend()
plt.grid()
```

thetaは概ね0.5度以下の位置にある。これは、学習済みの地点での精度を評価する事ができる。

phiを見てみると、学習した点以外の角度についても評価することができる。おおむね2度以下の精度で推論ができている事がわかる。

前回と比較すると、ピークが高くなったことで、thetaの推論の精度が向上した。

```python
print("phiの平均誤差(度)",np.mean(np.abs(diff_p)))
print("thetaの平均誤差（度）",np.mean(np.abs(diff_t)))
```
