```python
import tensorflow as tf
import numpy as np

# Allocation GPU hardware ID = 0
physical_devices = tf.config.list_physical_devices('GPU')
print(physical_devices)
tf.config.set_visible_devices(physical_devices[1], 'GPU')
```

```python
data_shape =(61, 61, 150,1)

model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Conv3D(16, 2, input_shape=(data_shape), activation='relu', strides=2))
# model.add(tf.keras.layers.MaxPool3D(1, 2))
model.add(tf.keras.layers.BatchNormalization())

model.add(tf.keras.layers.Conv3D(32, 2, activation='relu', strides=2))
# model.add(tf.keras.layers.MaxPool3D(1, 2))
model.add(tf.keras.layers.BatchNormalization())

model.add(tf.keras.layers.Conv3D(64, 2, activation='relu', strides=2))
# model.add(tf.keras.layers.MaxPool3D(1, 2))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.2))

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(361, activation='softmax'))

model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer=tf.keras.optimizers.SGD(0.2))

model.build()
model.summary()
```

```python
x = np.load("lern_data.npz")
y = np.load("lern_label.npz")

x = x["arr_0"]
x = np.expand_dims(x, axis=4)
y = y["arr_0"]

rand = np.array([np.random.randint(361) for i in range(15)])

print(rand)

x_test = [x[i] for i in rand]
y_test = [y[i] for i in rand]

x_train = x
y_train = y

# x_train = np.delete(x_train, rand, 0)
# y_train = np.delete(y_train, rand, 0)

x_train = np.array(x_train).astype('float32')
y_train = np.array(y_train).astype('float32')

x_test = np.array(x_test).astype('float32')
y_test = np.array(y_test).astype('float32')

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
batch_size = 1400
model.fit(x_train, y_train, batch_size=batch_size,
    epochs=900, validation_data=(x_test,y_test))
```

```python
y_pred = model.predict(x_test)
```

kerasのmodel.sequential ではネットワークのレイヤーを分岐させることはできない。

調べてみたところ、custom layer を使用することで分岐させることができそう。

~~~
class ResnetIdentityBlock(tf.keras.Model):
  def __init__(self, kernel_size, filters):
    super(ResnetIdentityBlock, self).__init__(name='')
    filters1, filters2, filters3 = filters

    self.conv2a = tf.keras.layers.Conv2D(filters1, (1, 1))
    self.bn2a = tf.keras.layers.BatchNormalization()

    self.conv2b = tf.keras.layers.Conv2D(filters2, kernel_size, padding='same')
    self.bn2b = tf.keras.layers.BatchNormalization()

    self.conv2c = tf.keras.layers.Conv2D(filters3, (1, 1))
    self.bn2c = tf.keras.layers.BatchNormalization()

  def call(self, input_tensor, training=False):
    x = self.conv2a(input_tensor)
    x = self.bn2a(x, training=training)
    x = tf.nn.relu(x)

    x = self.conv2b(x)
    x = self.bn2b(x, training=training)
    x = tf.nn.relu(x)

    x = self.conv2c(x)
    x = self.bn2c(x, training=training)

    x += input_tensor
    return tf.nn.relu(x)
